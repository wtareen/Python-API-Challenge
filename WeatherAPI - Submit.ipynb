```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from citipy import citipy
from config import api_key
from scipy.stats import linregress

```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
city = "sauda"

# Build query URL
query_url = url + "appid=" + api_key + "&q=" + city
response = requests.get(query_url).json()
response
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities_1 = []



# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

print(lat_lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name 
    
    # If the city is unique, then add it to a our cities list
    if city not in cities_1:
        cities_1.append(city)
# Print the city count to confirm sufficient count
len(cities_1)
```

```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="
query_url
```

```python
lat = []
lng=[]
temp = []
humidity = []
windspeed = []
cloudiness = []
final_cities = []
country = []
dt = []
# Loop through the list of cities and perform a request for data on each

for index,city in enumerate(cities_1):
    try:
        response = requests.get(query_url + city).json()
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        temp.append(response["main"]["temp_max"])
        humidity.append(response["main"]["humidity"])
        windspeed.append(response["wind"]["speed"])
        cloudiness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        dt.append(response["dt"])

        final_cities.append(city)
        print("city number is: " +str(index)+ ","" city name is : " +city )
        # print(f"Cityname: {index}|{city}")
#         time.sleep(15)
    except:
#         pass

        print("This is an error message!")

   
```

```python
weather_dict = {"Cities" : final_cities,
            "Country": country,
                "Date":dt,
               "Latitude" : lat,
               "Longitude":lng,
               "Temp" : temp,
               "Humidity": humidity,
               "Windspeed" : windspeed,
               "Cloudiness": cloudiness,
               }
weather_df = pd.DataFrame(weather_dict)
weather_df
#weather_df.to_csv("weathercsv.csv",index = False)
```

# Linear Regression

```python
# Create a function to create Linear Regression plots
def linearplot(x_values, y_values, ytitle, coordinates):
    
    # Run regresson on southern hemisphere
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
    regress_values = x_values * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

    # Plot
    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r-")
    plt.annotate(line_eq,coordinates,fontsize=15,color="red")
    plt.xlabel('Latitude')
    plt.ylabel(ytitle)
    print(f"The r-squared is: {rvalue}")
    plt.show()
```

```python
# Create Northern and Southern Hemisphere DataFrames
n_df = weather_df.loc[(weather_df["Latitude"] >= 0)]
s_df = weather_df.loc[(weather_df["Latitude"] < 0)]
```

# For Northern Hemisphere - Temperature model indicates that the model explains low variability of the response data around its mean.  The model fits the data well. There is high likelihood of future temperature predictions falling within the predicted outcome.

```python
# Linear regression on Northern Hemisphere - Temperature
x_values = n_df["Latitude"]
y_values = n_df["Temp"]
linearplot(x_values, y_values, 'Max Temp',(10,-20))
```
